```python
def RemoveDuplicate(path):
    with open(path, 'r', encoding='utf-8') as f:
        lines = filter(None, (line.rstrip() for line in f))
        print("Process start ...")
        setf = set()
        for line in lines:
            setf.add(line)
        wf = open(path, 'w', encoding='utf-8')
        for text in setf:
            wf.write(text+'\n')
        wf.close()
    print('Done!')
```

```python
import string
def text_preprocess(s):
    return s.lower()
```

```python
import re

# Contractions mapping
contractions_map = {
    "can't": "can not",
    "could've": "could have",
    "couldn't": "could not",
    "didn't": "did not",
    "hadn't": "had not",
    "hasn't": "has not",
    "haven't": "have not",
    "he'd": "he would",
    "he'll": "he will",
    "he's": "he is",
    "i'd": "I would",
    "i'll": "I will",
    "i'm": "I am",
    "i've": "I have",
    "it'd": "it would",
    "it'll": "it will",
    "it's": "it is",
    "let's": "let us",
    "mustn't": "must not",
    "shan't": "shall not",
    "she'd": "she would",
    "she'll": "she will",
    "she's": "she is",
    "should've": "should have",
    "shouldn't": "should not",
    "that's": "that is",
    "there's": "there is",
    "they'd": "they would",
    "they'll": "they will",
    "they're": "they are",
    "they've": "they have",
    "we'd": "we would",
    "we're": "we are",
    "we've": "we have",
    "weren't": "were not",
    "what'll": "what will",
    "what're": "what are",
    "what's": "what is",
    "what've": "what have",
    "where's": "where is",
    "who'd": "who would",
    "who'll": "who will",
    "who're": "who are",
    "who's": "who is",
    "who've": "who have",
    "won't": "will not",
    "would've": "would have",
    "wouldn't": "would not",
    "you'd": "you would",
    "you'll": "you will",
    "you're": "you are",
    "you've": "you have"
}

def expand_contractions(sentence):
    contractions_pattern = re.compile('({})'.format('|'.join(contractions_map.keys())), 
                                      flags=re.IGNORECASE | re.DOTALL)
    
    def expand_match(contraction):
        match = contraction.group(0)
        expanded_match = contractions_map.get(match.lower())
        return expanded_match

    expanded_sentence = contractions_pattern.sub(expand_match, sentence)
    return expanded_sentence
```

```python
import string
from underthesea import word_tokenize
from pygoogletranslation import Translator
src = 'en'
des = 'vi'
data_name = 'data.txt'
translator = Translator()
def clean_data(path, remove_pattern = 'Ngày'):
    with open(path, 'r', encoding='utf-8') as f:
        lines = filter(None, (line.rstrip() for line in f))
        print("Process start ...")
        for line in lines:
            line = line.strip()
            if line.startswith(remove_pattern):
                continue
            else:
                s_sentence = line
                try:
                    s_sentence = expand_contractions(s_sentence)
                    d_sentence = translator.translate(s_sentence, src=src, dest=des).text
                    d_sentence = word_tokenize(d_sentence[:-1] if d_sentence[-1] in ' .,?!_' else d_sentence, format='text')
                    s_sentence = s_sentence[:-1] if s_sentence[-1] in ' .,?!_' else s_sentence
                    
                    s_sentence = text_preprocess(s_sentence)
                    d_sentence = text_preprocess(d_sentence)

                    combine = s_sentence+'\t'+d_sentence+'\n'
                    wf = open(data_name, 'a', encoding='utf-8')
                    wf.write(combine)
                    wf.close()
                except Exception as e:
                    print(f' + got error {e} in "{line}" skip this, we will continue!')
                    continue
    print('All done!')
```

```python
clean_data('./data/data_raw_raw.txt')
```

## split to source and target file

```python
def split2(datapath, sourcepath, targetpath):
    with open(datapath, 'r', encoding='utf-8') as f:
        wfs = open(sourcepath, 'a', encoding='utf-8')
        wft = open(targetpath, 'a', encoding='utf-8')
        lines = filter(None, (line.rstrip() for line in f))
        for line in lines:
            line = line.strip()
            lists = line.split('\t')
            source_text, target_text = lists[0][:-1], lists[1][:-1]

            wfs.write(source_text+'\n')

            wft.write(target_text+'\n')

        wfs.close()
        wft.close()
```

```python
split2('data.txt', './data/Corpus/source.txt', './data/Corpus/target.txt')
```
